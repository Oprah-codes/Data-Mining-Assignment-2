```python
import pandas as pd
dataset=pd.read_csv(r"C:\Users\Wow\Downloads\US_Top_50_Universities_2026.csv")
```

```python
dataset.head()
```

```python
import numpy as np
```

```python
dataset_encoded=pd.get_dummies(dataset,columns=["University_Name","Institution_Type","State"])
```

```python
numeric_cols=dataset_encoded.select_dtypes(include=np.number).columns
```

```python
basket=dataset_encoded.astype(bool)
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
frequent_items=apriori(basket,min_support=0.1, use_colnames=True)
```

```python
rules=association_rules(frequent_items,metric="lift",min_threshold=1)
rules.head()
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
rules.sort_values(by="confidence",ascending=False)
```
